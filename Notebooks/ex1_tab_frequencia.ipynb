```python
import pandas as pd

# Leitura do arquivo CSV com o pandas
# O encoding também foi ajustado
usuarios = pd.read_csv("Usuarios.csv", sep=",", decimal=".", encoding="windows-1252")
print("Nomes das colunas:")
print(usuarios.columns)
print("-" * 30)
print("Estrutura do DataFrame:")
usuarios.info()
print("-" * 30)
print("Resumo estatístico:")
print(usuarios.describe())
```

```python
freq = usuarios['grau_instrucao'].value_counts()
print(freq)
```

```python
# Para obter a frequência relativa com .value_counts(), usamos o argumento normalize=True
freq_rel = usuarios['grau_instrucao'].value_counts(normalize=True)
print(freq_rel)
```

```python
# Simplesmente multiplicamos a frequência relativa por 100
p_freq_rel = usuarios['grau_instrucao'].value_counts(normalize=True) * 100
print(p_freq_rel)
```

```python
# Em pandas, podemos adicionar uma nova linha ao final da Series
freq['Total'] = freq.sum()
freq_rel['Total'] = freq_rel.sum()
p_freq_rel['Total'] = p_freq_rel.sum()
# O nome da última linha da 'freq' já foi definido como 'Total',
# não sendo necessário renomear as outras, pois faremos a junção em um novo DataFrame.
```

```python
tabela_final = pd.DataFrame({
    'freq': freq,
    'freq_rel': freq_rel,
    'p_freq_rel': p_freq_rel
})
print(tabela_final)
```
