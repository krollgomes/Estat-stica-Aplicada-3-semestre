```python
import pandas as pd
```

```python
# Carregar os dados
df = pd.read_csv("/content/dados_pacientes.csv")
```

```python
# Selecionar a coluna de interesse
peso = df['peso(kg)']
```

```python
# Cálculos
media = peso.mean()
mediana = peso.median()
moda = peso.mode()[0]  # .mode() retorna uma série, pegamos o primeiro valor
```

```python
print(f"Média: {media:.2f} kg")
print(f"Mediana: {mediana} kg")
print(f"Moda: {moda} kg")
```
