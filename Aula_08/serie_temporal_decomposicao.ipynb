```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

```

```python
# 1. Carregar os dados
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Aplicar log para transformar o modelo em aditivo
serie_log = np.log(df['Passengers'])
```

```python
# 2. Realizar a decomposição total para obter os componentes
result = seasonal_decompose(serie_log, model='additive', period=12)

tendencia = result.trend
sazonalidade = result.seasonal
residuo = result.resid
```

```python
# 3. PROCESSO DE REMOÇÃO PASSO A PASSO
# Passo A: Remover a Sazonalidade (Série Dessazonalizada)
# Na prática: Mostra o valor real sem o peso das "altas temporadas"
serie_sem_sazonalidade = serie_log - sazonalidade

# Passo B: Remover a Tendência da série já dessazonalizada
# Na prática: O que sobra é apenas o Ruído (Irregularidade) [cite: 53]
serie_so_ruido = serie_sem_sazonalidade - tendencia
```

```python
# 4. Visualização do Processo
fig, axes = plt.subplots(3, 1, figsize=(10, 12), sharex=True)

# Gráfico 1: Série Original
axes[0].plot(serie_log, color='blue', label='Série Original (log)')
axes[0].set_title("1. Série Original (Mistura de todos os componentes)")
axes[0].legend()

# Gráfico 2: Série sem Sazonalidade
axes[1].plot(serie_sem_sazonalidade, color='green', label='Original - Sazonalidade')
axes[1].set_title("2. Série Dessazonalizada (Apenas Tendência + Ruído)")
axes[1].legend()

# Gráfico 3: Apenas o Ruído
axes[2].plot(serie_so_ruido, color='red', label='Dessazonalizada - Tendência')
axes[2].set_title("3. Apenas Ruído (O que não é explicado por padrão nenhum)")
axes[2].legend()

plt.tight_layout()
plt.show()
```
